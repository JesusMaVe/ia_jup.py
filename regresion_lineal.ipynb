```python
#Librerias
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Datos de trabajo
x = np.array([1,2,3,4,5]).reshape(-1, 1) # Reestructuracion para scikit-learn
y = np.array([2,4,5,4,5])
```

```python
#Creacion de la instancia del modelo
model = LinearRegression()
```

```python
# Ajuste del modelo
model.fit(x, y)
```

```python
#obtencion de los parametro del modelo
print(f"Pendiente (B1): {model.coef_[0]}")
print(f"Intercepto  (B0): {model.intercept_}")
```

```python
# Uso del modelo para predecir x 
y_pred = model.predict(x)
print(f"Predicted values: ", y_pred)

rmse2 = sqrt(mean_squared_error(y, y_pred))
print("RMSE:", round(rmse2, 3))
```

```python
# Uso del modelo para apredecir con x1
x1 = np.array([8, 3, 5, 9, 2]).reshape(-1, 1)
y_pred = model.predict(x1)
print("Predicted values:", y_pred)
```

```python
plt.figure(figsize=(10, 6))
plt.title("Modelo de regresion lineal", size=16)
plt.scatter(x, y)
plt.plot(x, y_pred, c="red")
plt.show()
```
